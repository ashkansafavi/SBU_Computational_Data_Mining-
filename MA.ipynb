```python
import numpy as np

def crossDetc(x,f,g):
    crossUP=[]
    crossDOWN=[]
    idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
    for i in idx[5:]:
        if(g[i - 1] > f[i - 1]):
            crossUP.append(i)
        else : 
            crossDOWN.append(i)
            
    plt.plot(x,f)
    plt.plot(x,g)
    plt.plot(x[crossUP], f[crossUP], 'o')
    plt.plot(x[crossDOWN], f[crossDOWN], 'ro')

    plt.show()
```

```python
def goldenCrossDetc(x,f,g):
    goldenCross=[]
    crossDOWN=[]
    idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
    for i in idx[5:]:
        if(f[i+1] > f[i]):
            goldenCross.append(i)
        else : 
            crossDOWN.append(i)
            
    plt.plot(x,f)
    plt.plot(x,g)
    plt.plot(x[goldenCross], g[goldenCross], 'o')

    plt.show()
```

```python
from sklearn.linear_model import LinearRegression

def regPlot(x,y):
    model = LinearRegression().fit(x, y)
    
    #y_pred = model.intercept_ + model.coef_ * x
    
    y_pred = model.predict(x)

    plt.plot(y,x,'o')

    plt.plot(y_pred,x)
    plt.plot(y_pred,x - np.std(x)) 
    plt.plot(y_pred,x + np.std(x))

 


```

```python
regPlot(np.array(ds['high'].tail(n=10)).reshape(-1, 1),y = np.array(ds['openTime'].tail(n=10)).reshape(-1, 1))

    
```
